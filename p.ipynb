```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings("ignore", category=FutureWarning)
# Load the dataset
df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")
```

```python

# Load the dataset


# 1. Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
# Fill NaNs manually with inplace=True

df['mocodes'].fillna(df['mocodes'].mode()[0], inplace=True)
df['vict_sex'].fillna(df['vict_sex'].mode()[0], inplace=True)
df['vict_descent'].fillna(df['vict_descent'].mode()[0], inplace=True)
df['premis_cd'].fillna(df['premis_cd'].mean(), inplace=True)
df['weapon_used_cd'].fillna(df['weapon_used_cd'].mean(), inplace=True)
df['status'].fillna(df['status'].mode()[0], inplace=True)
df['crm_cd_1'].fillna(df['crm_cd_1'].mode()[0], inplace=True)
df['crm_cd_2'].fillna(df['crm_cd_2'].mode()[0], inplace=True)
df['crm_cd_3'].fillna(df['crm_cd_3'].mode()[0], inplace=True)
df['crm_cd_4'].fillna(df['crm_cd_4'].mode()[0], inplace=True)
df['cross_street'].fillna(df['cross_street'].mode()[0], inplace=True)
df['weapon_desc'].fillna(df['weapon_desc'].mode()[0], inplace=True)
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
warnings.filterwarnings("ignore", category=UserWarning)

# Load cleaned data
##df = pd.read_csv("Crime_Data_Cleaned.csv")

# Convert 'date_occ' to datetime if not already
df['date_occ'] = pd.to_datetime(df['date_occ'], errors='coerce')

# Extract year and month for grouping
df['year'] = df['date_occ'].dt.year
df['month'] = df['date_occ'].dt.to_period('M')

# Group by month to get number of crimes per month
monthly_trend = df.groupby('month').size().reset_index(name='crime_count')

# Convert 'month' back to datetime for plotting
monthly_trend['month'] = monthly_trend['month'].dt.to_timestamp()

# Plot the trend
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_trend, x='month', y='crime_count', marker='o')
plt.title("Monthly Crime Trend in Los Angeles")
plt.xlabel("Month")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python

# Load cleaned data
#df = pd.read_csv("Crime_Data_Cleaned.csv")

# Count top 10 crime types
top_crimes = df['crm_cd_desc'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(y=top_crimes.index, x=top_crimes.values, palette="Reds_r")
plt.title("Top 10 Most Common Crimes in Los Angeles")
plt.xlabel("Number of Cases")
plt.ylabel("Crime Type")
plt.tight_layout()
plt.show()
```

```python


# Load cleaned dataset
#df = pd.read_csv("Crime_Data_Cleaned.csv")

# Group by crime type and weapon, count occurrences
crime_weapon = df.groupby(['crm_cd_desc', 'weapon_desc']).size().reset_index(name='count')

# Get top 10 crime types
top_crimes = df['crm_cd_desc'].value_counts().head(10).index

# Filter for top crime types only
crime_weapon_top = crime_weapon[crime_weapon['crm_cd_desc'].isin(top_crimes)]

# Pivot for heatmap
pivot_table = crime_weapon_top.pivot_table(index='crm_cd_desc', columns='weapon_desc', values='count', fill_value=0)

# Plot heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, cmap='Reds', linewidths=0.5, linecolor='gray')
plt.title("🔥 Crime Type vs Weapon Used")
plt.xlabel("Weapon Description")
plt.ylabel("Crime Type")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


# Load cleaned data
#df = pd.read_csv("Crime_Data_Cleaned.csv")

# Make sure 'time_occ' is properly formatted
df['time_occ'] = df['time_occ'].astype(str).str.zfill(4)
df['hour'] = df['time_occ'].str[:2].astype(int)

# Group by hour and area
hour_area = df.groupby(['hour', 'area_name']).size().reset_index(name='crime_count')

# Pivot for heatmap
pivot_table = hour_area.pivot(index='area_name', columns='hour', values='crime_count').fillna(0)

# Plot heatmap
plt.figure(figsize=(16, 8))
sns.heatmap(pivot_table, cmap='YlOrRd', linewidths=0.5, linecolor='gray')
plt.title("📍 Crime Density by Hour and Area")
plt.xlabel("Hour of Day (0–23)")
plt.ylabel("Area Name")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

# Load cleaned dataset
#df = pd.read_csv("Crime_Data_Cleaned.csv")

# -------- 1. Crime count by Victim Sex --------
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='vict_sex', order=df['vict_sex'].value_counts().index, palette='Set2')
plt.title("Crime Distribution by Victim Sex")
plt.xlabel("Victim Sex")
plt.ylabel("Crime Count")
plt.tight_layout()
plt.show()

# -------- 2. Crime by Victim Age Group --------
# Create age bins
bins = [0, 12, 18, 30, 45, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Mid-age', 'Senior']
df['age_group'] = pd.cut(df['vict_age'], bins=bins, labels=labels, right=False)

# Plot
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='age_group', palette='coolwarm', order=labels)
plt.title("Crime Distribution by Victim Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Victims")
plt.tight_layout()
plt.show()

# -------- 3. Crime by Victim Descent --------
top_descents = df['vict_descent'].value_counts().head(10)
plt.figure(figsize=(8, 6))
sns.barplot(y=top_descents.index, x=top_descents.values, palette='viridis')
plt.title("Top 10 Victim Descent Groups in Crime Reports")
plt.xlabel("Number of Crimes")
plt.ylabel("Victim Descent")
plt.tight_layout()
plt.show()
```
